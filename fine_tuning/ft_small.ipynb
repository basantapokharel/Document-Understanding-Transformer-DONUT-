```python
from datasets import load_dataset

# Load dataset from Hugging Face Hub
dataset = load_dataset("katanaml-org/invoices-donut-data-v1")
dataset
```

```python
from transformers import DonutProcessor, VisionEncoderDecoderModel
import torch

# Load processor (image & tokenizer) and model (vision encoder + text decoder)


processor = DonutProcessor.from_pretrained("naver-clova-ix/donut-small", use_fast=True)
model = VisionEncoderDecoderModel.from_pretrained("naver-clova-ix/donut-small")

# Check if GPU is available, then move model to GPU for faster training
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print(f"Using device: {device}")
```
